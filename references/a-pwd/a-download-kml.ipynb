#### a-download-kml

 - file that downloads kmls from PWD website
 - you can just run all cells straight through

```python
import geopandas as gpd
import fiona
from time import sleep
import pandas as pd
```

```python
#list of all existing road IDs
roadID = pd.read_csv('roads.csv')['Road ID']
```

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

dfs = []
for i in roadID:
    url = f"https://mis.pwduk.in/im/roadKml/{i}.kml"
    sleep(1)
    try:
        print(f"Get road {i}")
        df = gpd.read_file(url, driver='KML')
        df['id'] = i
        dfs.append(df)
    except:
        print(f"KML for road id {i} does not exist")
        
    
all_dfs = pd.concat(dfs)
```

```python
all_dfs.to_file("/home/jovyan/work/data/b-roads/pwd-roads.gpkg", driver="GPKG")
```
